```python
import numpy as np  
import scipy.stats as stats
import statsmodels.api as sm
import matplotlib.pyplot as plt
import pylab

# Numpy Q-Q Plot Example 01:
measurements = np.random.normal(loc = 20, scale = 5, size=100) 
stats.probplot(measurements, dist="norm", plot=pylab)
pylab.show()
```

```python
# Numpy Q-Q Plot Example 02: 
test = np.random.normal(0,1, 1000)
sm.qqplot(test, line='45')
pylab.show()
```

```python
# Dow-Jones Industrial Average for Years 1996-2005
# These Data multiplied by 100
# Q-Q Plot for Ex.4.23-(a):
dowJones = np.array([-0.6, 3.1, 25.3, -16.8, -7.1, -6.2, 16.1, 25.2, 22.6, 26.0])
stats.probplot(dowJones, dist="norm", plot=pylab)
# sm.qqplot(dowJones, line='45')
pylab.show()
```

```python
# The World's 10 Largest Companies from www.Forbes.com partially based on Forbes The Forbes Global 2000
# April 18, 2005
sales = np.array([108.28, 152.36, 95.04, 65.45, 62.97, 263.99, 265.19, 285.06, 92.01, 165.68])
profits = np.array([17.05, 16.59, 10.91, 14.14, 9.52, 25.33, 18.54, 15.73, 8.10, 11.13])
assets = np.array([1484.10, 750.33, 766.42, 1110.46, 1031.29, 195.26, 193.83, 191.11, 1175.16, 211.15])

fig = plt.figure()
ax = fig.add_subplot(111)
stats.probplot(sales, dist="norm", plot=pylab)
stats.probplot(profits, dist="norm", plot=pylab)
stats.probplot(assets, dist="norm", plot=pylab)
#sm.qqplot(assets, line='45')
ax.set_title("Q-Q Plots of the World's 10 Largest Companies")
pylab.show()
```
